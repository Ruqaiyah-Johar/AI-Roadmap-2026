```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
print("shape of x:")
print(x.shape)
print("shape of y:")
print(y.shape)
```

```python
print("Name of feature are:")
print(iris.feature_names)
print("Name of target are:")
print(iris.target_names)
```

### Basic statistics

```python
print("Mean of each feature:")
print(np.mean(x,axis=0))
print("Standard deviation of each feature:")
print(np.std(x,axis=0))
```

### First 5 examples of iris dataset

```python
print("First 5 rows of x:")
print(x[:5])
print("First 5 labels:")
print(y[:5])
```

### Plot1- Iris:feature 0 vs 1

```python
import matplotlib.pyplot as plt

plt.scatter(x[:, 0], x[:, 1], c=y, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Iris: feature 0 vs 1")
plt.show()
```

### challenge 1- column stats

```python
#finding min in each row
print("min value in each row:")
print(np.min(x,axis=0))
print(np.min(x,axis=1))

#finding max in each row
print("max value in each row:")
print(np.max(x,axis=0))
print(np.max(x,axis=1))
```

### challenge 2 -filter by condition


```python
petal_length = x[:, 2]          
mask = petal_length > 4.0       
x_long = x[mask]               
y_long = y[mask]               

print("Number of long-petal flowers:")
print(x_long.shape[0])
print("First 5 rows:\n")
print(x_long[:5])
print("Their labels:")
print(y_long[:5])
```
